```python
from tkinter import *
from tkinter import messagebox
from tkinter import ttk
from tkinter import Text
import json
    

class studentApp:
    data = 'student_data.json'
    student_list=[]
    
    def __init__(self,root):
        self.root = root
        self.menu()
        self.lay_data()
        
  #================================================MENU===============================================================#      
    
    def menu(self):
        self.tree = ttk.Treeview(self.root, columns=("Mã SV", "Họ và Tên", "Năm sinh", "Quê quán", "Ngành", "Lớp"))
        self.tree.heading("#0", text="STT")
        self.tree.heading("#1", text="Mã SV")
        self.tree.heading("#2", text="Họ và Tên")
        self.tree.heading("#3", text="Năm sinh")
        self.tree.heading("#4", text="Quê quán")
        self.tree.heading("#5", text="Ngành")
        self.tree.heading("#6", text="Lớp")
        self.tree.pack()
        
        self.menu_label = Label(self.root, text="===== Ứng dụng Quản lý Sinh viên =====")
        self.menu_label.pack()

        self.xem_button = Button(self.root, text="Xem danh sách sinh viên", command=self.xem_danh_sach)
        self.xem_button.pack()
    
        self.button_add = Button(self.root, text="Thêm sinh viên mới", command=self.them_sinh_vien)
        self.button_add.pack()

        self.button_update = Button(self.root, text="Cập nhật thông tin sinh viên", command=self.update_student)
        self.button_update.pack()

        self.button_delete = Button(self.root, text="Xóa thông tin sinh viên", command=self.del_student)
        self.button_delete.pack()

        self.button_search = Button(self.root, text="Tìm kiếm sinh viên", command=self.search_student)
        self.button_search.pack()

        self.button_sort = Button(self.root, text="Sắp xếp danh sách sinh viên", command=self.sort_student_list)
        self.button_sort.pack()
        
        self.button_save = Button(self.root, text= "Lưu trữ", command = self.save_data)
        self.button_save.pack()

        self.button_exit = Button(self.root, text="Thoát", command=self.exit)
        self.button_exit.pack()
        
      #=================================================IMPORT/EXPORT DATA====================================================#      
        
    def lay_data(self):
        global student_list
        try:
            with open('student_data.json', "r", encoding="utf-8") as myfile:
                self.student_list = json.load(myfile)
        except:
            messagebox.showwarning(title="Lỗi", message="Lỗi đọc file")
            student_list = []
    
    def save_data(self):
        try:
            with open('student_data.json', "w", encoding="utf-8") as myfile:
                json.dump(self.student_list, myfile, indent = 4)
        except:
            messagebox.showwarning("Lỗi", "Lỗi lưu file")
            
    #=======================================================EXIT=============================================================#
    
    def exit(self):
        self.root.destroy()
        
    #=====================================================VIEW DATA==========================================================#  
    def xem_danh_sach(self):
        self.tree.delete(*self.tree.get_children())
        print(self.student_list)
        for sv in self.student_list:
            self.tree.insert("", "end", values=(sv["Mã SV"], sv["Họ và Tên"], sv["Năm sinh"], sv["Quê Quán"], sv["Ngành"], sv["Lớp"]))
     
    #=================================================ADD DATA=====================================================#  
    def them_sinh_vien_2(self,masinhvien,hoten,namsinh,quequan,nganh,lop):
        try:
            masinhvien = int(masinhvien)
            hoten = str(hoten)
            namsinh = int(namsinh)
            quequan = str(quequan)
            nganh = str(nganh)
            lop = str(lop)
            
            if masinhvien and hoten and namsinh and quequan and nganh and lop  is not None:
                self.student_list.append({
                "Mã SV": int(masinhvien),
                'Họ và Tên': hoten,
                "Năm sinh": int(namsinh),
                "Quê Quán": quequan,
                "Ngành": nganh,
                "Lớp": lop
                })
                messagebox.showinfo("Thông báo", "Thêm sinh viên thành công!")
                self.xem_danh_sach()
            else:
                messagebox.showwarning("Lỗi", "Vui lòng nhập đầy đủ thông tin")
            
        except:
            messagebox.showwarning("Lỗi", "Lỗi Định dạng dữ liệu nhập vào")
            
            
    def them_sinh_vien(self):
        top = Toplevel(self.root)
        top.title("Thêm Sinh Viên")
        top.geometry("300x400")
        
        msv_var = StringVar()
        name_var = StringVar()
        dob_var = StringVar()
        town_var = StringVar()
        major_var = StringVar()
        class_var = StringVar()
        
        lbl_add_student_msv = Label(top,text="Mã Sinh viên")
        lbl_add_student_msv.pack()
        ety_add_student_msv = Entry(top, textvariable=msv_var)
        ety_add_student_msv.pack()
        
        lbl_add_student_name = Label(top,text="Họ và Tên")
        lbl_add_student_name.pack()
        ety_add_student_name = Entry(top, textvariable=name_var)
        ety_add_student_name.pack()
        
        lbl_add_student_dob = Label(top,text="Năm Sinh")
        lbl_add_student_dob.pack()
        ety_add_student_dob = Entry(top, textvariable=dob_var)
        ety_add_student_dob.pack()
        
        lbl_add_student_town = Label(top,text="Quê Quán")
        lbl_add_student_town.pack()
        ety_add_student_town = Entry(top, textvariable=town_var)
        ety_add_student_town.pack()
        
        lbl_add_student_major = Label(top,text="Ngành")
        lbl_add_student_major.pack()
        ety_add_student_major = Entry(top, textvariable=major_var)
        ety_add_student_major.pack()
        
        lbl_add_student_class = Label(top,text="Lớp")
        lbl_add_student_class.pack()
        ety_add_student_class = Entry(top, textvariable=class_var)
        ety_add_student_class.pack()
        
        btn_add_student = Button(top,text="Thêm",command=lambda: self.them_sinh_vien_2(msv_var.get(),
                                                                               name_var.get(),
                                                                               dob_var.get(),
                                                                               town_var.get(),
                                                                               major_var.get(),
                                                                               class_var.get()))
        btn_add_student.pack()
    
    
    #=================================================UPDATE DATA=====================================================#  
    
    def update_student(self):
        top = Toplevel(self.root)
        top.title("Câp Nhật Thông Tin Sinh Viên")
        top.geometry("300x300")
        
        update_student_id = StringVar()
        
        Label(top, text="Nhập MSSV Cần Cập Nhật").pack()
        Entry(top, textvariable = update_student_id).pack()
        Button(top,text="Tìm", command=lambda: self.update_student_action(update_student_id)).pack()
        
        
    def update_student_action(self,update_student_id):
        for sv in self.student_list:
            if int(update_student_id.get()) == sv['Mã SV']:
                
                top = Toplevel(self.root)
                top.title("Câp Nhật Thông Tin Sinh Viên")
                top.geometry("300x300")
                
                sv_id = StringVar()
                sv_id.set(sv['Mã SV'])
                Label(top,text="Mã Sinh Viên").pack()
                Entry(top,textvariable=sv_id).pack()
                
                sv_name = StringVar()
                sv_name.set(sv['Họ và Tên'])
                Label(top,text="Họ và Tên").pack()
                Entry(top,textvariable=sv_name).pack()
                
                sv_dob = StringVar()
                sv_dob.set(sv['Năm sinh'])
                Label(top,text="Năm Sinh").pack()
                Entry(top,textvariable=sv_dob).pack()
                
                sv_town = StringVar()
                sv_town.set(sv['Quê Quán'])
                Label(top,text="Quê Quán").pack()
                Entry(top,textvariable=sv_town).pack()
                
                sv_major = StringVar()
                sv_major.set(sv['Ngành'])
                Label(top,text="Ngành").pack()
                Entry(top,textvariable=sv_major).pack()
                
                sv_cls = StringVar()
                sv_cls.set(sv['Lớp'])
                Label(top,text="Lớp").pack()
                Entry(top,textvariable=sv_cls).pack()
                
                
                btn_update_student = Button(top,text="Cập Nhật",command=lambda: self.update_student_action_2(update_student_id.get(),sv_id.get(),
                                                                                                      sv_name.get(),
                                                                                                      sv_dob.get(),
                                                                                                      sv_town.get(),
                                                                                                      sv_major.get(),
                                                                                                      sv_cls.get())).pack()
                
    def update_student_action_2(self,update_student_id,masinhvien,hoten,namsinh,quequan,nganh,lop):
        try:
            masinhvien = int(masinhvien)
            hoten = str(hoten)
            namsinh = int(namsinh)
            quequan = str(quequan)
            nganh = str(nganh)
            lop = lop
        except:
            None
        for sv in self.student_list:
            if int(update_student_id) == sv['Mã SV']:
                sv['Mã SV'] = masinhvien
                sv['Họ và Tên'] = hoten
                sv['Năm sinh'] = namsinh
                sv['Quê Quán'] = quequan
                sv['Ngành'] = nganh
                sv['Lớp'] = lop
                self.xem_danh_sach()
                messagebox.showinfo("Thông Báo","Cập Nhật Thành Công")
                

     #=================================================DELETE DATA=====================================================#     
    def del_student(self):
        top = Toplevel(self.root)
        top.title("Xoá Thông Tin Sinh Viên Trong Danh Sách")
        top.geometry("300x300")
        
        del_student_id = StringVar()
        
        Label(top, text="Nhập MSSV Cần Xoá").pack()
        Entry(top, textvariable = del_student_id).pack()
        Button(top,text="Xoá", command=lambda: self.del_student_action(del_student_id)).pack()
        
    def del_student_action(self,del_student_id):
        for sv in self.student_list:
            if del_student_id.get() == sv['Mã SV']:
                self.student_list.remove(sv)
                messagebox.showinfo("Thông báo", "Xoá Sinh Viên Thành Công")
                self.xem_danh_sach()
                
    #=================================================SEARCH DATA=====================================================#  
                
    def search_student(self):
        top = Toplevel(self.root)
        top.title("Tìm Thông Tin Sinh Viên")
        top.geometry("300x300")
        
        search_student_id = StringVar()
        Label(top, text="Nhập MSSV Cần Tìm").pack()
        Entry(top, textvariable = search_student_id).pack()
        Button(top,text="Tìm",command=lambda: self.search_student_action(search_student_id)).pack()
    
    def search_student_action(self,search_student_id):
        if not search_student_id.get():
            print("Lỗi")
        else:
            for sv in self.student_list:
                if int(search_student_id.get()) == sv['Mã SV']:
                    self.tree.delete(*self.tree.get_children())
                    self.tree.insert("", "end", values=(sv["Mã SV"], sv["Họ và Tên"], sv["Năm sinh"], sv["Quê Quán"], sv["Ngành"], sv["Lớp"]))
    
    #=================================================SORT DATA=====================================================#  
    def sort_student_list(self):
        self.student_list.sort(key=lambda x: x['Mã SV'])
        self.xem_danh_sach()
        
def lenhchay():
    main = Tk()
    main.title("Student Management System")
    command = studentApp(main)
    main.mainloop()
    
lenhchay()
```
